```python
import pandas as pd
pd.set_option('display.max_rows', 100)
df_produto = pd.read_csv('../data/produto.csv')
df_produto
```

```python
df_produto['primeiro_nome'] = df_produto['descItem'].apply(lambda x: x.lower().split(" ")[0])
```

```python
df_produto.sort_values(by=['vlPreco', 'descItem'],
                       ascending=[False, True])

df_produto.drop_duplicates(subset=['primeiro_nome', 'vlPreco'], keep='first')
```
