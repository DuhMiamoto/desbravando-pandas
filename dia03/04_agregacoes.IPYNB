```python
import pandas as pd

import numpy as np

df_produto = pd.read_csv('../data/produto.csv')
df_produto
```

```python
df_produto['vlPreco'].describe()
```

```python
df_produto['vlinflacao'] = (df_produto['vlPreco']*1.09).round(2)
df_produto.describe()
```

```python
df_produto['descItemPrimeiro'] = df_produto['descItem'].apply(lambda x: x.lower().split(" ")[0])
freq_items = pd.value_counts(df_produto['descItemPrimeiro'])
```

```python
df_produto.groupby(by=['descItemPrimeiro'],)['vlPreco'].mean()
df_produto.groupby(by=['descItemPrimeiro'],).describe()
```

```python
df_produto.groupby(by=['descItemPrimeiro'],).agg({'vlPreco': ['mean', 'min', 'max']})
(df_produto.groupby(by=['descItemPrimeiro'],)[['vlPreco', 'vlinflacao']]
                            .agg(['mean', 'max', 'min',]))
```
