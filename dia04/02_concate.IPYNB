```python
import pandas as pd

df_mods = pd.DataFrame(
    {
    'nome':['Kozato', 'sky', 'Nah', 'Karlla'],
     'idade': [25, 34, 33, 35]
     }
)

df_mods
```

```python
df_subs = pd.DataFrame(
    {   'nome':['sky', 'Mari', 'Bell', 'Maltrapiro', 'Matheus'],
        'idade': [34, 19, 19, 35, 26],
        'meses_subs':[24, 1, 1, 6, 1]
    }
)

df_subs
```

```python
(pd.concat([df_mods, df_subs])#empilha os dados
   .sort_values(['nome', 'meses_subs'])#coloca em ordem 
   .drop_duplicates(subset='nome')#apaga as duplicatas com base no nome pegando o primeiro
   .fillna(0))#transforma NAN em 0
```
